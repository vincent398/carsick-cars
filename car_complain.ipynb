```python
#数据加载
import pandas as pd

df = pd.read_csv('./car_complain.csv')
df
#print(df)
#df.to_excel('./car_complain.xlsx', index=False)
#导出为excel后缀应为.xlsx，False首字母需大写

#df = df.drop('problem', axis=1).join(df.problem.str.get_dummies(','))
#print(df)
```

```python
# 0代表没有，1代表存在
df = df.drop('problem', axis=1).join(df.problem.str.get_dummies(','))
df
```

按照brand统计 投诉总数，不同problem类型的总数<br>
按照投诉总数进行排序<br>
按照指定的problem类型进行排序


```python
#数据清洗，将别名合并
#替换函数f(x)
def f(x):
    x = x.replace('一汽-大众','一汽大众')
    return x

df['brand'] = df['brand'].apply(f)

#聚合，以brand分组，聚合id数量
result = df.groupby(['brand'])['id'].agg(['count'])
result
```

```python
#标签从0，1，2，……，7第七个A11开始
tags = df.columns[7:]
tags
```

```python
result2 = df.groupby(['brand'])[tags].agg(['sum'])
result2
```

```python
result2 = result.merge(result2, left_index=True, right_index=True, how='left')
result2
```

```python
#brand目前为索引列，将brand复原为正常列，reset索引列
result2.reset_index(inplace=True)
result2
```

```python
result2.to_csv("./result.csv")
```

```python
#按照count，从大到小进行排序
result2 = result2.sort_values('count', ascending=False)
result2
```

```python
result2.to_csv("./result.csv")
```
